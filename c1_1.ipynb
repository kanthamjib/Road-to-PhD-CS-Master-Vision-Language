#### Note: Intro to NLP 

(Ref: https://www.geeksforgeeks.org/nlp/introduction-to-natural-language-processing-nlp/)

NLP is the study that create models and frameworks to help computer understand human languages. 
The NLP application, such as, Large Language Models (LLM), Optical Character Recognition (OCR), Vision-Language Models (VLM), Speech-to-text (ASR), and Text-to-Speech (TTS).

The following steps are the flow of NLP generally applied:

1. Text Input/Speech Input (Through Speech Recognition) -> This is to obtain the text data.

2. Preprocessing (Several Cleaning Methods) -> This is where "Tokenization" happens.

3. Language Analysis -> To analyze grammar structures, word relationships, context, and semantic meanings in text.

4. Text Representation (The heart of NLP) -> This is where "Word Embeddings" revolutionized NLP by representing words as dense vectors that capture semantic similarity. Later contextual embeddings further improved NLP by modeling words based on surrounding context.

5. Model Training -> When every words become vector, let's neural network learn everything.

6. Output Generation

Therefore, today we will learn **Tokenization and Word Embeddings** or we call the **“Foundation of Text Representation”**.
